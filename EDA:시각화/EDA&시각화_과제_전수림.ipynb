### 2) 결측치 확인

- #### <font color='orange'> Q2. 각 컬럼의 결측치를 확인하고, 결측치가 많은 순서대로 결측치 비율과 함께 표로 나타내보세요.
<font color='orange'> (결측치 0개인 컬럼은 나타내지 마세요)

```python
#각 컬럼의 결측치 확인
df.isnull().sum()
```

```python
#결측치 비율
nullPer = (df.isnull().sum() / len(df)) * 100
nullPer
```

```python
nullNum = df.isnull().sum()
nullCheckDf = pd.concat([nullNum, nullPer], axis=1, keys=['nullNum', 'nullPer'])
```

```python
nullCheckDf = nullCheckDf[nullCheckDf['nullNum']>0]
nullCheckDf = nullCheckDf.sort_values('nullNum', ascending=False)
nullCheckDf
```

- #### <font color='orange'> Q3. 결측치가 있는 컬럼과 결측치 비율을 보기쉽게 그래프로 나타내 보세요.

```python
plt.barh(nullCheckDf.index, nullCheckDf['nullPer'])
```
